```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/drive/MyDrive/Software/grass.jpg')
```

```python
cv2_imshow(img)
```

```python
#convert to grayscale
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#display new image
cv2_imshow(img_gray)
```

```python
#resize an image
img_resize = cv2.resize(img,(500,500))

#display resized image
cv2_imshow(img_resize)
```

```python
#flip an image
img_flip = cv2.flip(img,0)

#display an image
cv2_imshow(img_flip)
```
